```python
import numpy as np
```

```python
# Kod je preuzet sa: https://github.com/giotto-ai/giotto-tda/blob/master/examples/data/generate_datasets.py
def make_point_clouds(n_samples_per_shape: int, n_points: int, noise: float):
    circle_point_clouds = [
        np.asarray(
            [
                [np.sin(t) + noise * (np.random.rand(1)[0] - 0.5), np.cos(t) + noise * (np.random.rand(1)[0] - 0.5), 0]
                for t in range((n_points ** 2))
            ]
        )
        for kk in range(n_samples_per_shape)
    ]

    circle_labels = np.zeros(n_samples_per_shape)

    sphere_point_clouds = [
        np.asarray(
            [
                [
                    np.cos(s) * np.cos(t) + noise * (np.random.rand(1)[0] - 0.5),
                    np.cos(s) * np.sin(t) + noise * (np.random.rand(1)[0] - 0.5),
                    np.sin(s) + noise * (np.random.rand(1)[0] - 0.5),
                ]
                for t in range(n_points)
                for s in range(n_points)
            ]
        )
        for kk in range(n_samples_per_shape)
    ]

    sphere_labels = np.ones(n_samples_per_shape)

    torus_point_clouds = [
        np.asarray(
            [
                [
                    (2 + np.cos(s)) * np.cos(t) + noise * (np.random.rand(1)[0] - 0.5),
                    (2 + np.cos(s)) * np.sin(t) + noise * (np.random.rand(1)[0] - 0.5),
                    np.sin(s) + noise * (np.random.rand(1)[0] - 0.5),
                ]
                for t in range(n_points)
                for s in range(n_points)
            ]
        )
        for kk in range(n_samples_per_shape)
    ]

    torus_labels = 2 * np.ones(n_samples_per_shape)

    point_clouds = np.concatenate((circle_point_clouds, sphere_point_clouds, torus_point_clouds))
    labels = np.concatenate((circle_labels, sphere_labels, torus_labels))

    return point_clouds, labels
```

```python
n_samples_per_class = 20
points_per_dimension = 15 # svaki cloud ce imati ppd^2 tacaka
noise = 0.3
point_clouds, labels = make_point_clouds(n_samples_per_class, points_per_dimension, noise)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import ipywidgets as widgets
from ipywidgets import interact
```

```python
circle_idx = 0
sphere_idx = n_samples_per_class
torus_idx = (2 * n_samples_per_class)

fig = go.FigureWidget(
    make_subplots(
        rows=1, cols=3,
        specs=[[{'type': 'scene'}, {'type': 'scene'}, {'type': 'scene'}]],
        subplot_titles=('Circle', 'Sphere', 'Torus')
    )
)

fig.add_trace(
    go.Scatter3d(
        x=point_clouds[circle_idx, :, 0],
        y=point_clouds[circle_idx, :, 1],
        z=point_clouds[circle_idx, :, 2],
        mode='markers',
        marker=dict(size=3, color='crimson', opacity=0.7),
        name='Circle'
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter3d(
        x=point_clouds[sphere_idx, :, 0],
        y=point_clouds[sphere_idx, :, 1],
        z=point_clouds[sphere_idx, :, 2],
        mode='markers',
        marker=dict(size=3, color='forestgreen', opacity=0.7),
        name='Sphere'
    ),
    row=1, col=2
)

fig.add_trace(
    go.Scatter3d(
        x=point_clouds[torus_idx, :, 0],
        y=point_clouds[torus_idx, :, 1],
        z=point_clouds[torus_idx, :, 2],
        mode='markers',
        marker=dict(size=3, color='royalblue', opacity=0.7),
        name='Torus'
    ),
    row=1, col=3 
)

fig.update_layout(
    height=550, 
    width=1500, 
    title_text=f"Sample ID: {0}",
    title_font_size=18,
    showlegend=False
)

fig.update_scenes(xaxis=dict(title='X Axis', range=[-1.5, 1.5]),
                  yaxis=dict(title='Y Axis', range=[-1.5, 1.5]),
                  zaxis=dict(title='Z Axis', range=[-0.5, 0.5]), row=1, col=1)

fig.update_scenes(xaxis=dict(title='X Axis', range=[-1.5, 1.5]),
                  yaxis=dict(title='Y Axis', range=[-1.5, 1.5]),
                  zaxis=dict(title='Z Axis', range=[-1.5, 1.5]), row=1, col=2)

fig.update_scenes(xaxis=dict(title='X Axis', range=[-3.5, 3.5]),
                  yaxis=dict(title='Y Axis', range=[-3.5, 3.5]),
                  zaxis=dict(title='Z Axis', range=[-1.5, 1.5]), row=1, col=3)


def update_all_plots(offset):
    c_idx = offset
    s_idx = n_samples_per_class + offset
    t_idx = (2 * n_samples_per_class) + offset
    
    with fig.batch_update():
        fig.data[0].x = point_clouds[c_idx, :, 0]
        fig.data[0].y = point_clouds[c_idx, :, 1]
        fig.data[0].z = point_clouds[c_idx, :, 2]
        
        fig.data[1].x = point_clouds[s_idx, :, 0]
        fig.data[1].y = point_clouds[s_idx, :, 1]
        fig.data[1].z = point_clouds[s_idx, :, 2]
        
        fig.data[2].x = point_clouds[t_idx, :, 0]
        fig.data[2].y = point_clouds[t_idx, :, 1]
        fig.data[2].z = point_clouds[t_idx, :, 2]
        
        fig.update_layout(title_text=f"Sample ID: {offset}")

slider = widgets.IntSlider(
    value=0,
    min=0,
    max=n_samples_per_class-1,
    step=1,
    description='Sample ID:',
    continuous_update=False
)

interact(update_all_plots, offset=slider)
display(fig)
```

```python
from gtda.homology import VietorisRipsPersistence
from gtda.plotting import plot_diagram
from gtda.diagrams import PersistenceEntropy
```

```python
VR = VietorisRipsPersistence(homology_dimensions=[0, 1, 2])
diagrams = VR.fit_transform(point_clouds)

PE = PersistenceEntropy()
features = PE.fit_transform(diagrams)
```

```python
def update_persistence_plot(diagram_idx):
    single_diagram = diagrams[diagram_idx]
    
    fig = plot_diagram(single_diagram)

    name_from_index = lambda x: "Circle" if x < n_samples_per_class else "Sphere" if x < 2*n_samples_per_class else "Torus"
    
    fig.update_layout(
        title=f"Persistence Diagram {diagram_idx} ({name_from_index(diagram_idx)})",
        width=800,
        height=600
    )
    
    fig.show()

slider = widgets.IntSlider(
    value=0,
    min=0,
    max=len(diagrams)-1,
    step=1,
    description='Diagram ID:',
    continuous_update=False
)

interact(update_persistence_plot, diagram_idx=slider);
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(features, labels)
model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_valid, y_valid)
```

```python
features_no_h0 = features[:, 1:]
X_train, X_valid, y_train, y_valid = train_test_split(features_no_h0, labels)
model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_valid, y_valid)
```
